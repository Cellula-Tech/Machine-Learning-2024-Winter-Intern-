```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report , confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Step 2: Data Preprocessing
# Assume 'target' is the name of your target column
df = pd.read_csv(r"C:\Users\Khaled\Downloads\first inten project.csv")
X = df.drop(columns=['booking status'])  # Features
y = df['booking status']  # Target variable
```

```python
# Convert categorical variables if necessary
X = pd.get_dummies(X, drop_first=True)
```

```python

from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=4,
                           n_informative=2, n_redundant=0,
                           random_state=42, shuffle=False)
clf = RandomForestClassifier(max_depth=2, random_state=42)
clf.fit(X, y)
```

```python

# Step 3: Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 5: Evaluate the model accuracy before feature selection
y_pred =clf.predict(X_test)
accuracy_before = accuracy_score(y_test, y_pred)
print(f'Accuracy before feature selection: {accuracy_before:.2f}')
```

```python

feature_names = [f'feature_{i}' for i in range(X.shape[1])]
X = pd.DataFrame(X, columns=feature_names)
y = pd.Series(y)

# Step 6: Perform feature selection using Random Forest
importances = clf.feature_importances_
feature_names = X.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
print(feature_importance_df)
```

```python

# Step 7: Select top N features (example selecting top 10 features)
top_features = feature_importance_df['Feature'][:10].values
feature_indices = [X.columns.get_loc(feature) for feature in top_features]
X_train_selected = X_train[:, feature_indices]
X_test_selected = X_test[:, feature_indices]
```

```python
# Step 8: Train a new Random Forest model (after feature selection)
rf_selected = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf_selected.fit(X_train_selected, y_train)
```

```python
# Step 9: Evaluate the model accuracy after feature selection
y_pred = rf_selected.predict(X_test_selected)
accuracy_after = accuracy_score(y_test, y_pred)
print(f'Accuracy after feature selection: {accuracy_after:.2f}')
```

```python
# Cross-validation scores
from sklearn.model_selection import train_test_split, cross_val_score
cv_scores = cross_val_score(rf_selected, X_train_selected, y_train, cv=5)
print(f'Cross-validation accuracy: {cv_scores.mean():.2f} Â± {cv_scores.std():.2f}')
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
rf = RandomForestClassifier(random_state=42)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#for encoding
from sklearn.preprocessing import LabelEncoder

#for train test splitting
from sklearn.model_selection import train_test_split

#for decision tree object
from sklearn.tree import DecisionTreeClassifier

#for checking testing results
from sklearn.metrics import classification_report, confusion_matrix

#for visualizing tree 
from sklearn.tree import plot_tree
```

```python
X_loan.shape,y_loan.shape
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_loan, y_loan, test_size=0.3, random_state=42)
```

```python
# Create a decision tree classifier
model = DecisionTreeClassifier(criterion='gini',max_depth=8, random_state=42)
```

```python
# Train the decision tree
model.fit(X_train, y_train)
```

```python
# Make predictions on the testing set
y_pred = model.predict(X_test)

y_pred
```

```python
feature_names = [f'feature_{i}' for i in range(X.shape[1])]
X = pd.DataFrame(X_loan, columns=feature_names)
y = pd.Series(y_loan)

# Step 6: Perform feature selection using Random Forest
importances = model.feature_importances_
feature_names = X_loan.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
print(feature_importance_df)


# Feature selection
from sklearn.feature_selection import SelectFromModel
# Feature selection
selector = SelectFromModel(model, threshold='mean', prefit=True)
X_train_selected = selector.transform(X_train.values)  # Convert to NumPy array
X_test_selected = selector.transform(X_test.values)    # Convert to NumPy array
```

```python
# Calculate accuracy after feature selection
accuracy_after = accuracy_score(y_test, y_pred_selected)
print(f'Accuracy after feature selection: {accuracy_after:.4f}')
print(classification_report(y_test, y_pred_selected))
```
